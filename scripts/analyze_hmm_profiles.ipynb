```python
import pandas as pd
import os
import shutil
```

```python
df_full = pd.read_csv('report_hmms.csv')
df_full
```

```python
df_full['Model type'].value_counts()
```

```python
df_full_length_model = df_full[df_full['Model type'] == 'full-length']
df_full_length_model
```

```python
model_ids = set(df_full[df_full['Model type'] == 'full-length']['Model_ID'])
output_dir = 'full_length_hmms'
os.makedirs(output_dir, exist_ok=True)

for root, dirs, files in os.walk('picornaviridae'):
    for file in files:
        if file.endswith('.hmm'):
            model_id = file.split('.')[0]
            if model_id in model_ids:
                src = os.path.join(root, file)
                dst = os.path.join(output_dir, file)
                shutil.copy2(src, dst)
```
