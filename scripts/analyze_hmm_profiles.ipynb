```python
import pandas as pd
import os
import shutil
```

```python
df = pd.read_csv('..\\data\\minion_DB_picornaviridae_hmm_profiles_metadata.csv')
#df
```

```python
df['Model type'].value_counts()
```

```python
df_full_length_model = df[df['Model type'] == 'full-length']
#df_full_length_model
```

```python
#model_ids = set(df[df['Model type'] == 'full-length']['Model_ID'])
#output_dir = '..\\data\\hmms\\full_length_hmms'

model_ids = set(df[df['Model type'] == 'short']['Model_ID'])
output_dir = '..\\data\\hmms\\short_length_hmms'

os.makedirs(output_dir, exist_ok=True)

for root, dirs, files in os.walk('..\\data\\hmms\\picornaviridae_all_hmms'):
    for file in files:
        if file.endswith('.hmm'):
            model_id = file.split('.')[0]
            if model_id in model_ids:
                src = os.path.join(root, file)
                dst = os.path.join(output_dir, file)
                shutil.copy2(src, dst)
```

```python
df_short_length_model = df[df['Model type'] == 'short']
#df_short_length_model
```

```python
viruses_path = '..\\data\\list_of_viruses_from_viral_zone.txt'
viruses_list = []
with open(viruses_path, 'r') as f:
    for line in f:
        viruses_list.append(line.strip())
```

```python
df_all_euk_full = pd.read_csv('..\\data\\hmms\\minion_all_eukaryotic\\Short\\report_hmms.csv')
df_all_euk_full = df_all_euk_full[['Model_ID', 'Taxon', 'Model type', 'Positive terms', 'Start coordinate', 'End coordinate', 'Length', '# of sequences', 'Genome type', 'Parent']]
df_all_euk_full = df_all_euk_full[df_all_euk_full['Genome type'] == 'ssRNA']
df_all_euk_full = df_all_euk_full[df_all_euk_full['Taxon'].isin(viruses_list)]
df_all_euk_full
```

```python
model_ids = set(df_all_euk_full['Model_ID'])
output_dir = '..\\data\\hmms\\minion_short_viral_zone'

os.makedirs(output_dir, exist_ok=True)

for root, dirs, files in os.walk('..\\data\\hmms\\minion_all_eukaryotic\\Short'):
    for file in files:
        if file in model_ids:
            src = os.path.join(root, file)
            dst = os.path.join(output_dir, file)
            shutil.copy2(src, dst)
```
