```python
import pandas as pd 
import numpy as np

df = pd.read_csv('../data/Astroviridae_16022026_full.csv')
df
```

```python
df = df[df['class'] == 'Mammalia']
df
```

```python
frames = ["1A", "1B", "2"]

for frame in frames:
    
    df[frame] = df[frame].replace("NA", np.nan)
    
    coords = df[frame].str.split("-", expand=True)
    
    start = pd.to_numeric(coords[0], errors="coerce")
    end = pd.to_numeric(coords[1], errors="coerce")

    df[f"{frame}_length"] = end - start #+ 1
    df
```

```python
from Bio import SeqIO

gb_file = "../data/Astroviridae_15102025.gb"
csv_file = "../data/Astroviridae_mammalia.csv"
output_file = "../data/Astroviridae_filtered.gb"

df = pd.read_csv(csv_file)

valid_ids = set(df["ID"].astype(str))

print("Accessions in table:", len(valid_ids))

kept_records = []

for record in SeqIO.parse(gb_file, "genbank"):
    #print(record.id, record.annotations.get("accessions"))
    accession = record.id.split('.')[0] 
    
    if accession in valid_ids:
        kept_records.append(record)

print("Records kept:", len(kept_records))

SeqIO.write(kept_records, output_file, "genbank")
```
