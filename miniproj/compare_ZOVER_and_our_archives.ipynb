```python
import pandas as pd
```

```python
df_zover = pd.read_excel('../ZOVER_all_bat_viruses.xlsx', header=1)
df_zover
```

```python
df_our = pd.read_csv('../sra_filtered_no_16S.csv', index_col=False)
df_our
```

```python
matching_rows = df_our[df_our['BioSample'].isin(df_zover['BioSample'])]
matching_rows
```

```python
matching_rows.to_csv('matching.csv')
```

```python
not_matching_rows = df_our[~df_our['BioSample'].isin(df_zover['BioSample'])]
not_matching_rows['Collection_Date'].value_counts().to_csv('not_matched_dates.csv')
```

```python
not_matching_rows.to_csv('not_matching.csv')
```
