```python
import pandas as pd
```

```python
df1 = pd.read_excel("../diamond_best_hits_our_profiles.xlsx")
df2 = pd.read_excel("../diamond_best_hits_not_our_profiles.xlsx")
```

```python
df1['compare_key'] = df1['qseqid'].str.split('|').str[-1].str.split('/').str[1:4].str.join('/')
df2['compare_key'] = df2['qseqid'].str.split('|').str[-1].str.split('/').str[1:4].str.join('/')

df1_only = df1[~df1['compare_key'].isin(df2['compare_key'])]
df2_only = df2[~df2['compare_key'].isin(df1['compare_key'])]

non_matching_rows = pd.concat([df1_only, df2_only])

non_matching_rows
```
