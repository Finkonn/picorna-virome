```python
import pandas as pd
import glob
import numpy as np

cols = [
    "contig", "score", "evalue", "pident", "length",
    "subject", "annotation", "taxid"
]

files = glob.glob("*.tsv")  # adjust path

dfs = []

for f in files:
    df = pd.read_csv(f, sep="\t", names=cols)
    
    df["sample"] = f.split("/")[-1].replace(".tsv", "")
    df["virus"] = df["annotation"].str.extract(r"\[(.*?)\]")
    df["protein"] = df["annotation"].str.extract(r"^(.*?)\s*\[")
    
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
```
